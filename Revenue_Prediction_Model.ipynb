#Analysing Data and Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Month_Value_1.csv")
df
```

```python
df.Period=pd.to_datetime(df.Period,format="%d.%m.%Y")
df.dtypes
```

```python
df.set_index('Period',inplace=True)
df
```

```python
df=df.asfreq('MS')
df.index
```

```python
df.dropna(axis=0,inplace=True)
```

```python
plt.plot(df.index,df.Revenue)
```

```python
df.head(30)
```

##Stationarity Check

```python
import statsmodels.tsa.stattools as sts
sts.adfuller(df.Revenue)
```

```python
df['revenue_returns']=df.Revenue.pct_change(1).mul(100)
df
```

```python
sts.adfuller(df.revenue_returns[1:])
```

##Seasonality Check

```python
from statsmodels.tsa.seasonal import seasonal_decompose
sd=seasonal_decompose(df.revenue_returns[1:],model="additive")
sd.plot()
```

##Splitting Data

```python
split_size=int(len(df)*0.8)
train=df.iloc[1:split_size]
test=df.iloc[split_size:]
train.shape,test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_scaled=sc.fit_transform(train)
test_scaled=sc.transform(test)
```

```python
train_scaled = pd.DataFrame(train_scaled, columns=train.columns, index=train.index)
test_scaled= pd.DataFrame(test_scaled, columns=test.columns, index=test.index)
```

#Auto-ARIMA Model

```python
from pmdarima.arima import auto_arima
auto_arima_model=auto_arima(train_scaled['revenue_returns'],trace=True,suppress_warnings=True)
auto_arima_model.summary()
```

```python
auto_arima_model.fit(train_scaled['revenue_returns'])
auto_arima_test_prediction = auto_arima_model.predict(n_periods=len(test_scaled))
```

```python
from sklearn.metrics import mean_squared_error
print("Root mean square error = "+str(np.sqrt(mean_squared_error(test_scaled['revenue_returns'], auto_arima_test_prediction))))
plt.plot(test_scaled.index,test_scaled.revenue_returns)
plt.plot(test_scaled.index,auto_arima_test_prediction)
plt.legend(['actual','predicted'])
```

#ARIMA

```python
import statsmodels.graphics.tsaplots as sgt
sgt.plot_acf(df.revenue_returns[1:], lags=15)
```

```python
sgt.plot_pacf(df.revenue_returns[1:],lags=15)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from statsmodels.tsa.arima.model import ARIMA
ARIMA_models={}
for p in range(0,6):
  for d in range(0,2):
    for q in range(0,7):
      order=(p,d,q)
      model=ARIMA(train_scaled['revenue_returns'],order=order)
      model_fit=model.fit()
      forecast=model_fit.forecast(steps=13)
      error=np.sqrt(mean_squared_error(test_scaled['revenue_returns'],forecast))
      ARIMA_models[order]=error
      print("ARIMA"+str(order)+" MSE = "+str(error))
```

```python
best_model_ARIMA=min(ARIMA_models,key=ARIMA_models.get)
print("Best model will be ",best_model_ARIMA," with an error of ",ARIMA_models[best_model_ARIMA])
```

```python
ARIMA_model=ARIMA(train_scaled['revenue_returns'],order=best_model_ARIMA)
ARIMA_model_fit=ARIMA_model.fit()
ARIMA_model_fit.summary()
```

```python
ARIMA_test_perdictions=ARIMA_model_fit.forecast(steps=13)
print(test_scaled['revenue_returns'].shape)
print(ARIMA_test_perdictions.shape)
```

```python
print("Root mean square error = "+str(np.sqrt(mean_squared_error(test_scaled['revenue_returns'], ARIMA_test_perdictions))))
plt.plot(test_scaled.index,test_scaled.revenue_returns)
plt.plot(test_scaled.index,ARIMA_test_perdictions)
plt.legend(['actual','predicted'])
```

#LSTM

```python
def df_to_X_y(df, window_size=3):
    X = []
    y = []
    date= df.index
    date=date.to_numpy()
    date=date[window_size:]
    np_data=df.to_numpy()
    for i in range(len(np_data)-window_size):
        row=[a for a in np_data[i:i+window_size]]
        X.append(row)
        label=np_data[i+window_size]
        y.append(label)

    return np.array(date),np.array(X),np.array(y)
```

```python
date_train,X_train,y_train=df_to_X_y(train_scaled['revenue_returns'])
date_test,X_test,y_test=df_to_X_y(test_scaled['revenue_returns'])
date_train.shape,X_train.shape,y_train.shape,date_test.shape,X_test.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import *

LSTM_model = Sequential()
LSTM_model.add(InputLayer((3,1)))
LSTM_model.add(LSTM(64))
LSTM_model.add(Dense(32, activation='relu'))
LSTM_model.add(Dense(32, activation='relu'))
LSTM_model.add(Dense(1))
LSTM_model.summary()
```

```python
LSTM_model.compile(loss='mse', optimizer=Adam(learning_rate=0.001),metrics=['mean_squared_error'])
LSTM_model.fit(X_train,y_train,epochs=500)
```

```python
LSTM_test_prediction = LSTM_model.predict(X_test).flatten()
print("Root mean square error = "+str(np.sqrt(mean_squared_error(y_test, LSTM_test_prediction))))
plt.plot(date_test, y_test)
plt.plot(date_test, LSTM_test_prediction)
plt.legend(['Actucal','Predicted'])
```
